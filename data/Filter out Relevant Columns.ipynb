# Filtering out Relevant Columns

We only want the `model` name, `x_score`, and `y_score` for the final matrix

```python
# Importing Librarires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in file
df = pd.read_excel('LLM Matrix.xlsx')
```

```python
# Have a look at the first few rows
df.head()
```

```python
# The headers are in the second row (iloc[1]) so rename:
df.rename(columns=df.iloc[1], inplace = True)
```

```python
# Drop of the first two rows
df.drop([0,1], inplace = True)
```

```python
# DataFrame with relevant coulmns only
relv_only = df[['model', 'x_score', 'y_score']]
```

```python
# Have a look
relv_only
```

```python
# Filter out rows with null scores (i.e. give df with both x and y scores)
relv_only.dropna(subset=['x_score', 'y_score'], inplace=True)
```

```python
# Have a look
relv_only
```

```python
# Save as file
relv_only.to_csv('relevant_cols_only.csv', encoding='utf-8', index=False)
```
